```python
def find_B(f):
    B = mat(f)
    cand_cols = identity_matrix(F,m)
    for i in range(m):
        tmp = B.augment(cand_cols[:,i])
        if rank(tmp) == rank(B) + 1:
            B = tmp
        if B.ncols() == m:
            break

    return B
```

```python
def poly(v):
    return sum([v[i]*x**i for i in range(len(v))])
```

```python
def poly_to_vec(p,d):
    return vector(Fext,[p.coefficient(i) for i in range(d+1)])
```

```python
def vec_vec_prod(u, v, Q):
    u_poly = poly(u)
    v_poly = poly(v)
    tmp = (u_poly * v_poly) % Q

    return poly_to_vec(tmp, len(u)-1)
```

```python
def encrypt(message, key):
    f,g,D,s = key
    
    u = random_vector(Fext,n)
    p = poly(u)
    R2 = random_matrix(F,w,n)
    e = f*R2
    v = vec_vec_prod(s,u, Q) + e + (message*g[0])
    ct = (u,v)

    return ct
```

```python
def decrypt(ct, key):
    f,g,D,s = key
    u,v = ct
    d = D[:,0]

    message = (d).T * mat(v - vec_vec_prod(s, u, Q))

    return message
```

```python
def prod_const(ct, m2):
    u,v = ct
    u_ = vec_vec_prod(u,m2,Q)
    v_ = vec_vec_prod(v,m2,Q)

    return (u_, v_)
```

```python
def sum_ct(ct1,ct2):
    u1,v1 = ct1
    u2,v2 = ct2

    return (u1 + u2, v1 + v2)
```
